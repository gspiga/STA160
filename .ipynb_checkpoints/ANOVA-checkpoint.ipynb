# ANOVA with Gender and Age Groups



```python
import numpy as np
import pandas as pd
import plotly.express as px
from IPython.display import display
```

```python
data = pd.read_csv("Data/data_clean.csv", index_col = [0])

#Creating Age Groups
data.loc[(data['age'] > 14) & (data['age'] <= 24), 'Age Group'] = 'Youth (15-24)'
data.loc[(data['age'] > 25) & (data['age'] <= 64), 'Age Group'] = 'Adults (25-64)'
data.loc[(data['age'] >= 65), 'Age Group'] = 'Seniors (65+)'

# Making Gender Categorical
data["gender"].replace({1:"Male", 2: "Female", 3: "Other"}, inplace = True)
display(data)
```

```python
display(px.box(data, x = 'Age Group', y = 'score', color = 'gender'))
pd.crosstab(data["score"], data["Age Group"])
```
